This dataset includes information on a variety of categories: information on creators, information about content pushed by the ad, information on the reach of the ad (views), populations targeted, geographical locations targeted, as well as creative features of an ad (url, call-to-action). All these categories describe the various components of an ad, and allow us to investigative the underlying differences between ads from year 2018 and years from 2019 in depth, which will be the focus in this analysis.

Now, it is important to note that the EDA in this project is broken down in two parts: before investigating patterns in years, it was key to have a deeper understanding of all values present in the dataset and take notes on values that "dont make sense" that may affect future analysis. ### Assessment of Missingness
For the NMAR column in my dataset, I claimed that missingness in EndDate column is NMAR. Investigating ads where EndDate was null, it was found that most of this missingness started to appear more as a StartDate approached the end of 2019. This could potentially be explained due to the fact that most ads end in either the same year or the following year, if an advertiser decided to continue this ad for an even longer period of time than a year, it could potentially explain the missingness. Another possibiliy that is found is studying how Snapchat Ad campaigns are designed. Snapchat pushes creators to create lifetime plans for their ad to be put for a long time (source linked in cells). Following this assumption we find that spending for ads where EndDate was null was severely lower than most ads. Therefore, the Snapchat Ad lifetime plans could be another reason the missingnesss in EndDate, making its missingness dependant on the value reported under Spend. For my MAR missingness assessment, I selected to test for MAR on columns CandidateBallotInformation and CreativeProporties. This analysis focuses on testing if CandidateBallotInformation (CBI) missingnes is due to CreativeProperties (CP) missingness. Logic behind analysis is the claim that if there was no 'call-to-action' (CP) property in the ad, there was most likely no CBI displayed. The set-up for this test is a permuation test (of 500 trials) using a test statistic of Total Variation Distance (TVD), setting a significance level of 0.01. To begin this test, an empirical distribution between both columns is created. It is important to note that since both of these columns have mostly missing values, some values CBI in ads not represented in both True/False CP missing rows. For instance, "Aebischer" in CBI was only present in rows were CP was not missing. Since the TVD calculation will drop these NaN values across the columns (cannot subtract NaN from real numbers), the empirical distribution is graphed dropping the CBI columns in the empirical distribution that have NaN values only for illustration purposes on graph. Following this, we obtain the observed TVD in from the calculated empirical distribution, which is 0.2637. After calculating the observed, we preceed to the permutation test: Shuffling CP column, calculating TVDs and storing it into a results list. Concluding this test, we observe a p-value of 0.002. Since 0.002 < 0.01 (alpha), we conclude that missingness in CBI column is in fact dependant on the CreativeProperties column. In order to find a column where CBI missiness was not dependant on, I chose to investigate the Regions(Included) column, which describes the specified regions where an ad will be displayed. Looking at this two columns, the intuition behind testing lack of dependency is not seeing a direct connection between missing CBI and missing Regions(Included) since choosing to specify a certain region could extend outside a candidate campaign: maybe a non-profit in that region targeting residents within that region, etc. This test follows the same set up as MAR dependecy test: TVD is the chosen test statistic for permutation test (of 500 trials), with a significance level of 0.01. We calculate the empirical distribution and calculate the observed TVD, 0.1307. Similarly to the MAR test, we see a lot of missingness the empirical distribution, CBI categories that are not present in both True/False missingness in Regions(Included) for certain rows, which calculating TVD will also drop these values automatically. Moving to the permutation tests and plotting a histogram using the shuffled TVDs results, we see a p-val of 0.476, which >> 0.01(alpha). Therefore, we can conclude that CBI missingness is not dependant on missingness in Regions(Included) column. Missingness among these columns is important in testing the overall features of these ads between year, which will be investigated in the hypothesis test. If there is missiness in a call-to-action that is due to missiness in another column that may not describe a feature, this may cause a potential bias on the test where a feature is missing due to another column instead of a creator just not including that specified feature.

### Hypothesis Test
For my hypothesis test, I chose to test if the number of features in ads from 2019 was higher than ads made in 2018. In order to set up this test, it is vital to specify such "features". Features in my analysis are columns in the dataset, ads, that imply a specific target. This target can include geographical targets, specifications targeting a population, as well as more "trivial" specifications, such as Interests. The chosen columns for the test were Regions (Included), Regions (Excluded), Electoral Districts (Included), Electoral Districts (Excluded), Radius Targeting (Included), Radius Targeting (Excluded), Metros (Included), Metros (Excluded), Postal Codes (Included), Postal Codes (Excluded), Location Categories (Included), Location Categories (Excluded), Segments, AdvancedDemographics, and CreativeProperties. The CandidateBallotInformation column was not considered for this test due to its dependency in missingess with the CreativeProperties column that was founding in the MAR missingness assessment. If the value in these columns contains a specific target (specific postal code where the ad will be displayed, specific Interest targeted, etc.), this will count as a "specified feature" or target. On the other side, if the value is null, the ad contained no specified target. With these details noted, we specify the null hypothesis to be: total number of specified features in ads from 2018 and ads from 2019 have the same distribution, any difference seen is due to chance. On the other side, the alternative hypothesis is: total number of specified features in ads from 2019 is higher than ads from 2018, therefore ads in 2019 are targeting a more specific audience. This test will use a permutation test of 1000 trials using a test statistic of TVD, with a significance level of 0.01. To begin the test, information in ads from the selected columns as well as the Year column will be copied into a dataframe called targets. Null values in this dataframe are transformed to 0, and non-null are transformed into 1. Next, we create a column called Total that will hold the sum across rows for every ad and therefore hold the total number of specified targets in each ad. In order to compute the observed TVD, a conditional distribution was created from Year on Total and plotted in order to illustrate the distributions of Total on both years of 2018 and 2019. Computing the test statistic, we get a TVD of 0.1016. Moving on to the permutations, the Year column in targets is shuffled in each of the 1000 trials, and TVD is calculated for each of these trials and stored in a results list. After simulation, results of TVDs under the null hypothesis are illustrated in a histogram as well as our observed value. Calculating the p-val based on results and observed, we get a p-val of 0.00. Since this p-val is << 0.01 (alpha), we reject the Null Hypothesis conluding that the the total number of specified features in both years come from two different distributions were the total of "targets" is higher in ads from 2019.

### Cleaning and EDA

#### Concatenating Data

```python
# in order to easily differentiate and compare date from two years, we are going to create a column 'Year' before concat.

ads18.insert (0, 'Year', 2018)
ads19.insert(0, 'Year', 2019)
```

```python
# now we can concadenate both data sets

ads = pd.concat([ads18, ads19])
ads.head(3)
```

#### Cleaning: Start/EndTime

```python
# convering to pd.DateTime

#'Z' at end of Start/EndTime means UTC timezone

ads.StartDate = pd.to_datetime(ads.StartDate)
ads.EndDate = pd.to_datetime(ads.EndDate)

ads.StartDate.dtypes, ads.EndDate.dtypes # should output datetime obj w timezone UTC
```

#### Filling up NaNs according to readme.

```python
# Gender column: If nan means ALL genders all targeted
ads.Gender = ads.Gender.fillna('ALL')

# Age Bracket: If nan means all ages
ads.AgeBracket = ads.AgeBracket.fillna('all')

# Interest column: If nan means no particular interest was targeted
ads.Interests = ads.Interests.fillna('None')
```

```python
# AdvancedDemographics: If nan means no 3rd party data segments were used
ads.AdvancedDemographics = ads.AdvancedDemographics.fillna('None')

# OsType columns: if nan means all were targeted
ads.OsType = ads.OsType.fillna('ALL')

# Language columns: if nan means no specific language was targeted
ads.Language = ads.Language.fillna('None') #NOTE: language needs some cleaning

# Target Connection Type column: If nan means no particular connection type was targeted
ads['Targeting Connection Type'] = ads['Targeting Connection Type'].fillna('None')

# Targeting Carrier (ISP) column: If nan means all carriers were targeted
ads['Targeting Carrier (ISP)'] = ads['Targeting Carrier (ISP)'].fillna('All')
```

#### Cleaning: AgeBrackets

```python
# NOTICE vals in Gender: '18-', '16-', '19-'...

strings = ads[ads.AgeBracket.str.match('[0-9]{2}-$')].AgeBracket.unique()
strings # we need to replace all these
```

```python
# going thru every string and replacing thru '-' with '+'
for s in strings:
    ads.AgeBracket = ads.AgeBracket.replace({s: s[:2]+'+'})
```

#### Insertion of Columns: Duration, MonthStart/End, DayStart/End.

```python
# lets insert a column with the days of duration the ads were up on snapchat.

ads.insert(8, 'Duration', (ads.EndDate - ads.StartDate))
```

```python
# adding column MonthEnd.
ads.insert(9, 'MonthEnd', ads.EndDate.dt.month)
```

#### Investigating MAR Missingness

Testing for MAR, set alpha (significance level) = 0.01

```python
df = ads.copy() # not edit ads
```

```python
# claim: if Candidate Ballot info is missing, it can explain why the "call to action" (Creative URL) category is missing...
# if the Candidate Ballot info was in fact missing, we infer that the add may NOT have a call to action.

# Both of these tests will be conducted using a permutation test of N=500 trials, using TVD as the test start,
# both with a significance level (alpha) of 0.01.
```

```python
# testing for MAR using the empirical dist of Creative Prop on Candidate Ballot Info column.

distr = (
    df
    .assign(is_null=df['CreativeProperties'].isnull())
    .pivot_table(index='is_null', columns='CandidateBallotInformation', aggfunc='size')
    .apply(lambda x:x / x.sum(), axis=1)
)

distr

# NOTE: When looking at the empirical distribution generated, we see that some CBI was not present for both dist of False and True
# missing values (Creative Properties), therefore we CANNOT use CBI that do have a nan in its respective columns ...  we drop these CBI
```

```python
# plotting final emp dist

distr.dropna(axis=1).T.plot(kind='bar', title='Distribution of CBI when Creative Properties is missing/non-missing');
# note: I am only going to plot the values to make illustration of both distribituons better, since distr has 71 columns where
# some rows contain NaN values that will be dropped anyway by TVD.

# Observations: for "Get Out the Vote", we see an extreme misproportion of missigness when Creative Properties is null and when it is not!
# We can also see very small proportions of Joe Biden, Julegryta, General Election within the False pop, while on the True pop we see
# many that are also missing
```

```python
###### PERMUTATION TEST: P(CBI missing | CP missing)

# Statistic for test Selected: TVD

# 1. Calculation observed tvd in observed empirical dist

obs = distr.diff().iloc[-1].abs().sum() / 2
obs
```

```python
# 2. Performing permutations: Shuffling CP column, calculate tvds throughout each shuffled emp dist

n_repetitions = 500

tvds = []

for _ in range(n_repetitions):
    
    # shuffle the gender column
    shuffled_col = (
        df['CandidateBallotInformation']
        .sample(replace=False, frac=1)
        .reset_index(drop=True)
    )
    
    # put them in a table
    shuffled = (
        df
        .assign(**{
            'CandidateBallotInformation': shuffled_col,
            'is_null': df['CreativeProperties'].isnull()
        })
    )
    
    # compute the tvd
    shuffled = (
        shuffled
        .pivot_table(index='is_null', columns='CandidateBallotInformation', aggfunc='size')
        .apply(lambda x:x / x.sum(), axis=1)
    )
    
    tvd = shuffled.diff().iloc[-1].abs().sum() / 2
    # add it to the list of results
    
    tvds.append(tvd)
```

```python
# 3. observing pval
pval = np.mean(tvds > obs)
pd.Series(tvds).plot(kind='hist', density=True, alpha=0.8, title='P(CBI missing | CP missing) p-value: %f' % pval)
plt.scatter(obs, 0, color='red', s=40);

# since p-val is 0.002 < alpha(0.01), we can determine that the CBI column missingness is dependant 
# on 'CreativeProperties' (call to action) column.
```

```python
# now ... lets find a column where CBI column missingness is NOT dependant.

# lets test if CBI column has a dependency to the column 'Regions (Included)' 

# we repeat the previous process, getting the empirical dist first based on claim P(CBI missing| Regions (Included) null/nonnull)

distr2 = (
    df
    .assign(is_null=df['Regions (Included)'].isnull())
    .pivot_table(index='is_null', columns='CandidateBallotInformation', aggfunc='size')
    .apply(lambda x:x / x.sum(), axis=1)
)

# nan vals will be dropped anyway when calculating tvd
distr2.dropna(axis=1).T.plot(kind='bar', title='Distribution of CBI when Regions (Included) is missing/non-missing');
```

```python
###### PERMUTATION TEST: P(CBI missing| Regions (Included) missing)

# Statistic for test Selected: TVD

# 1. Calculation observed tvd in observed empirical dist (distr2)

obs = distr2.diff().iloc[-1].abs().sum() / 2
obs
```

```python
# 2. Performing permutations: Shuffling CP column, calculating tvds each iteration

n_repetitions = 500

tvds = []

for _ in range(n_repetitions):
    
    # shuffle the gender column
    shuffled_col = (
        df['CandidateBallotInformation']
        .sample(replace=False, frac=1)
        .reset_index(drop=True)
    )
    
    # put them in a table
    shuffled = (
        df
        .assign(**{
            'CandidateBallotInformation': shuffled_col,
            'is_null': df['Regions (Included)'].isnull()
        })
    )
    
    # compute the tvd
    shuffled = (
        shuffled
        .pivot_table(index='is_null', columns='CandidateBallotInformation', aggfunc='size')
        .apply(lambda x:x / x.sum(), axis=1)
    )
    
    tvd = shuffled.diff().iloc[-1].abs().sum() / 2
    # add it to the list of results
    
    tvds.append(tvd)
```

```python
# 3. observing pval
pval = np.mean(tvds > obs)
pd.Series(tvds).plot(kind='hist', density=True, alpha=0.8, title='P(CBI missing| Regions (Included) missing) p-value: %f' % pval)
plt.scatter(obs, 0, color='red', s=40);
```

```python
# p-val is too high! 0.48, so we can conclude missingness in CBI column is NOT dependant on missingness in Regions (Included).
```

### Hypothesis Test

```python
# Claim: ads in 2019 have a higher number of specific features in their ads than in 2018

# Null Hyp: The distribution of total specified features among ads in 2019 is the same as among ads in 2018.
# The difference between the two samples is due to chance.

# Alt Hyp: The distributions of total features of the two groups are different.

# Test Stat: TVD, Test: Permutation Test of N=1000 trials, alpha=0.01
# ------------------------------------------------------------------------------------------------------------------------ #

# For this test, I am only going to include feature columns that have missing values with no meaning.
# Testing for a Gender specification wouldnt hold since not specifying a gender in an ad is considered to be targeting ALL genders, 
# same for Interests column, Targeting Carrier, etc..
# Therefore, we are ONLY including specifications that have NO meaning unless specified by ad creator, such as
# location specifications, call to action, postal codes targeted ...


# t will hold the columns that we are going to consider as specific features of an ad.

t = list(ads.columns)[20:32] + ['Segments', 'AdvancedDemographics', 'CreativeProperties']
t
```

```python
# grabbing feature columns with 'Year' from ads, these will be held in df targets.
targets = ads[['Year'] + t].copy()

targets = targets.set_index('Year') # setting index to 'Year'


# a null entry corresponds to ad NOT specifying a field for that feature, therefore we change null entries to 1, and null to 0.
targets = targets.isnull().replace({True : 1, False : 0})

# adding a 'total' column to keep track of the total number of specified features per row
targets['total'] = targets.sum(axis=1)

targets = targets.reset_index() # getting 'Year' out of index to shuffle rows easily.

targets.head()
```

```python
# lets look at the distributions first of 'total' by 'Year'

title='total of speficied features by year'

G = targets.groupby('Year')

ax = G.get_group(2018)['total'].value_counts(normalize=True).rename('2018').plot(kind='bar', alpha=0.3, legend=True, title=title, color='r')
G.get_group(2019)['total'].value_counts(normalize=True).rename('2019').plot(kind='bar', alpha=0.3, ax=ax, legend=True);

# are these distributions different ? this are just the counts distributions, we need to see the conditional distribution
# dependant on the counts for each year since we have a different number of ads from 2018 and 2019.
```

```python
# lets view the empirical distribution instead 

# getting feature coutns from total based on 'Year' column
feature_cnts = targets.pivot_table(index='total', columns='Year', aggfunc='size')
feature_cnts

feature_cnts.sum().to_frame().T # how many counts for each group do we have ?
```

```python
# getting conditional dist

# since we are diving by total respective of each group, the conditional distr will tell us if there is a significant difference
# between the counts of features from both years

cond_distr = feature_cnts.apply(lambda x: x / x.sum())
cond_distr
```

```python
# plotting conditional distribution

title='distribution of total # of features conditional on year'
cond_distr.plot(kind='barh', title=title);
```

```python
# test statistic for permutation test

def tvd_of_groups(df):
    cnts = df.pivot_table(index='total', columns='Year', aggfunc='size')
    distr = cnts.apply(lambda x: x / x.sum())   # normalized 
    return distr.diff(axis=1).iloc[:, -1].abs().sum() / 2  # TVD
```

```python
# for this test, we are going to use tvd in order to find if there is a significant difference between both distributions

# 1. getting the obs tvd from out conditional dist we get
obs = tvd_of_groups(targets)

obs
```

```python
# 2. simulation

N = 1000
tvds = []
for _ in range(N):
    
    # shuffling Year column and calculate the tvd for shuffled conditional distr
    s = targets['Year'].sample(frac=1, replace=False).reset_index(drop=True)
    shuffled = targets.loc[:, ['total']].assign(Year=s)
    
    tvds.append(tvd_of_groups(shuffled))

tvds = pd.Series(tvds)
```

```python
# 3. what is the pval ?

pval = (tvds >= obs).mean()

# plotting tvds under null and observed value in our data set
pd.Series(tvds).plot(kind='hist', density=True, alpha=0.8, title='p-value: %f' % pval)
plt.scatter(obs, 0, color='red', s=40);
```

```python
# pval is too low! pval < alpha, meaning we reject the null hypothesis:

# 'total' count of specified features from 2018 and 2019 do not come from the same distributions

ads['Electoral Districts (Included)'].unique()
```
